````python
def f(x):
  return 2*np.sin(x) - np.cos(2*x)

def df(x):
  return 2*np.cos(x) + 2*np.sin(2*x)
````

Прописал функции, чтобы быстро получать значения функции и производной

````python
  def beauty_derivative_plot(px):
    x = np.linspace(-2*math.pi, 2*math.pi, 500)
    py = round(f(px), 2)
    plt.figure(figsize=(10, 5))
    plt.plot(x, f(x), 'b', label="$f(x) = 2sin(x) - cos(2x)$")
    plt.plot(x, df(x), color = 'lightcoral', linestyle = '--', label="$f'(x) = 2cos(x) + 2sin(2x)$")
    plt.plot(px, py, "ko")
    x_tangent = np.linspace(px - 1, px + 1, 100)
    plt.plot(x_tangent, f(px) + df(px)*(x_tangent - px), "g-", label=f"Tangent at x = {px:.2f}")

    y_range = np.linspace(min(f(px),0), max(f(px), 0), 2)
    x_range = np.linspace(min(px, 0), max(px, 0), 2)
    plt.plot([px, px], y_range, color='purple', linestyle='--', linewidth=1)
    plt.plot(x_range, [py, py], color='purple', linestyle='--', linewidth=1)

    plt.axhline(y=0, color='black', linewidth=0.8)
    plt.axvline(x=0, color='black', linewidth=0.8)

    plt.annotate(f"f'({px:.2f}) = {round(df(px),2)}", xy=(px, py), xytext=(px - 1, py-2),
                           color="black", fontsize=10,
                           arrowprops = dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

    plt.title("Function, Derivative and Tangent Line")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.axis([-2*math.pi,2*math.pi, -4, 4])
    plt.grid(True)
    plt.legend(loc='upper right')
    plt.show()
    return
````

Построили функции и точку, в которой хотим увидеть касание.
Построили касательную в точке (px, py).
Дальше построили фиолетовые пунктирные линии, которые показывают позицию точки на плоскости, выделяем оси, рисуем красивую стрелку и печатаем название нашего графика, маштабируем, активируем сетку и легенду, располагая её в правом верхнем углу.

## Problem 2

The data in [populations.txt](https://drive.google.com/file/d/1Bq7qVzbZIOd058VobGE6iYjUcMFkbZMt/view?usp=drive_link) describes the populations of hares, lynxes and carrots in northern Canada during 20 years. Do the following exercises **without for-loops**.

````python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('populations.txt')

years = data[:, 0]
hares = data[:, 1]
lynxes = data[:, 2]
carrot = data[:, 3]
````

- Plot the population size of each species for each year on a single graph. Add legend and axis labels to the plot.

````python
# Построение графика
plt.figure(figsize = (15,5))
plt.plot(years, hares, label = "Hares")
plt.plot(years, lynxes, label = "Lynxes", color = "green")
plt.plot(years, carrot, label = "Carrot", color = "orange")

plt.title("Population size of each species for each year")
plt.legend(loc="upper right")
plt.grid(True)
plt.xticks(years)
plt.xlabel("Years")
plt.ylabel("Population sizes")
````

- Find which year each species had the largest population. Mark these years in the plot of populations.

````python
# Поиск момента времени наибольшей популяции каждого вида
year_tlp_hares = years[np.argmax(hares)]
year_tlp_lynxes = years[np.argmax(lynxes)]
year_tlp_carrot = years[np.argmax(carrot)]

print(f"Year with largest hares population: {max(hares), int(year_tlp_hares)}")
print(f"Year with largest lynxes population: {max(lynxes), int(year_tlp_lynxes)}")
print(f"Year with largest carrot population: {max(carrot), int(year_tlp_carrot)}")

plt.figure(figsize = (15,5))
# Построение выделенных точек
plt.plot(year_tlp_hares, max(hares), color = "blue", marker = "o", label = "TLP Of Hares")
plt.plot(year_tlp_lynxes, max(lynxes), color = "green", marker = "o", label = "TLP Of Lynxes")
plt.plot(year_tlp_carrot, max(carrot), color = "orange", marker = "o", label = "TLP Of Carrot")

# Построение графика
plt.plot(years, hares, label = "Hares")
plt.plot(years, lynxes, label = "Lynxes", color = "green")
plt.plot(years, carrot, label = "Carrot", color = "orange")

plt.title("Population size of each species for each year")
plt.legend(loc="upper right")
plt.grid(True)
plt.xticks(years)
plt.xlabel("Years")
plt.ylabel("Population sizes")
plt.show()
````

- Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each specis for each year in the period.

````python
# Вычисление долей
total_population = hares + lynxes # листы складываются поэлементно
hares_ratio = hares/total_population
lynxes_ratio = lynxes/total_population

# Построение столбчатых диаграмм
plt.figure(figsize = (15,5))
plt.bar(years, hares_ratio, label = "Hares", color = "blue")
plt.bar(years, lynxes_ratio, bottom = hares_ratio, label="Lynxes", color = "red") #Столбцы рысей ставятся на столбцы зайцев как дополнение до 1
plt.xticks(years)
plt.legend(loc = "upper right")
plt.show()
````

- Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Show each histogram in a separate plot. Arrange 3 plots in a row.

````python
# Вычисление среднего значения и стандартного отклонения
mean_hares = np.mean(hares)
std_hares = np.std(hares)

mean_lynxes = np.mean(lynxes)
std_lynxes = np.std(lynxes)

mean_carrot = np.mean(carrot)
std_carrot = np.std(carrot)

print(f'The mean value of hares:{round(mean_hares,2)} and standard deviation: {round(std_hares,2)}')
print(f'The mean value of lynxes:{round(mean_lynxes,2)} and standard deviation: {round(std_lynxes,2)}')
print(f'The mean value of carrot:{round(mean_carrot,2)} and standard deviation: {round(std_carrot,2)}')

plt.figure(figsize = (15,5))

# Построение гистограммы для зайцев:
plt.subplot(1,4,1)
plt.hist(hares, bins = 10, color = 'blue')
plt.axvline(mean_hares, color = 'red', linestyle = 'dashed', linewidth = 1)
plt.title('Histogram Of Hares Population')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.legend(['Mean', 'Hares'])

# Построение гистограммы для рыси:
plt.subplot(1,4,2)
plt.hist(lynxes, bins = 10, color = 'green')
plt.axvline(mean_lynxes, color = 'red', linestyle = 'dashed', linewidth = 1)
plt.title('Histogram Of Lynxes Population')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.legend(['Mean', 'Lynxes'])

# Построение гистограммы для моркови:
plt.subplot(1,4,3)
plt.hist(carrot, bins = 10, color = 'orange')
plt.axvline(mean_carrot, color = 'red', linestyle = 'dashed', linewidth = 1)
plt.title('Histogram Of Carrot Population')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.legend(['Mean', 'Carrot'])

plt.tight_layout()
plt.show()
````

- Plot the change in the hare population and in the lynx population for each year. Find the correlation coefficient between the both time series.

````python
hares_change = np.diff(hares)
lynxes_change = np.diff(lynxes)

plt.figure(figsize=(15, 6))

plt.plot(years[1:], hares_change, label='Change in Hare Population', marker='o', linestyle='-')
plt.plot(years[1:], lynxes_change, label='Change in Lynx Population', marker='o', linestyle='-')


plt.xlabel('Years')
plt.ylabel('Change in Population')
plt.title('Change in Hare and Lynx Populations Each Year')
plt.legend()
plt.xticks(years)
plt.grid(True)

# Вычисление коэффициента корреляции
corr_coef = np.corrcoef(hares_change, lynxes_change)[0,1]
print(f"Correlation coefficient between hare and lynx population changes: {corr_coef}")

plt.tight_layout()
plt.show()
````

- Assume the population of hares in 1920 is unknown. Suggest a way to estimate this value. Compare an estimated value
with the true value and print a ratio of the error to the true value.

````python
# Запоминаем индекс 1920 года, сохраняем число рысей в 1920 году
year_1920_index = np.where(years == 1920)[0][0]
true_hare_1920 = hares[year_1920_index]
hares[year_1920_index] = np.nan

# Вычисляем средний прирост за год
valid_years = years[~np.isnan(hares)] # Записали все года, в которые данные о зайцах валидные, т.е. без 1920
valid_hares = hares[~np.isnan(hares)]
growth_rates = np.diff(valid_hares) / valid_hares[:-1]  # Вычисляем прирост последовательно по элементам в долях
average_growth_rate = np.mean(growth_rates)

# Находим год, непосредственно предшествующий 1920 году, где была валидная популяция зайцев
if len(valid_years) > 0:
  year_before = valid_years[-1]
else:
  print("There are no years before 1920 to use to make the estimate. Exiting program.")
  exit()
# print(year_before)

#Находим популяцию зайцев в тот самый год
hare_before = valid_hares[valid_years == year_before][0]
# print(hare_before)

# Оцениваем популяцию в 1920 году, используя средний прирост
estimated_hare_1920 = hare_before * (1 + average_growth_rate)

# Вычисляем погрешность
error = abs(estimated_hare_1920 - true_hare_1920)
error_ratio = error / true_hare_1920

# Восстанавливаем исходное значение
hares[year_1920_index] = true_hare_1920

# Результаты
print(f"True Hare Population in 1920: {true_hare_1920:.2f}")
print(f"Estimated Hare Population in 1920: {estimated_hare_1920:.2f}")
print(f"Absolute Error: {error:.2f}")
print(f"Ratio of Error to True Value: {error_ratio:.2f}")
````
